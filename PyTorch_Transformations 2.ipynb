```python
import os
import torch
from torchvision.transforms import v2
from PIL import Image
from torchvision.utils import save_image
from torchvision.io import read_image
from torchvision.transforms import transforms
```

```python
transforms1 = v2.Compose([
    v2.ToPILImage(),
    v2.RandomRotation(degrees=45),
    v2.ToTensor(),
])

transforms2 = v2.Compose([
    v2.ToPILImage(),
    v2.RandomPerspective(distortion_scale=0.6, p=1.0),
    v2.ToTensor(),
])
```

```python

file_names = os.listdir("/content/images")
filtered_file_names = [file_name for file_name in file_names if file_name.endswith(".jpg")]
count = 1
# I want to create 5 different random images for each original image.
for i in range(5):
  for file_name in filtered_file_names:
    file_name_without_extension, extension = os.path.splitext(file_name)
    img_tensor = read_image(f"/content/images/{file_name_without_extension}{extension}")
    # run transform RandomRotation
    img_transformed1 = transforms1(img_tensor)
    save_image(img_transformed1, f"/content/images/{file_name_without_extension}_RandomRotation_{count}{extension}")
    # run transform RandomPerspective
    img_transformed2 = transforms2(img_tensor)
    save_image(img_transformed2, f"/content/images/{file_name_without_extension}_RandomPerspective_{count}{extension}")
    count += 1
```
